###Question 1

```python
import pandas as pd
import numpy as np
```

```python
try:
  Energy = pd.read_excel("/content/Energy Indicators.xls", header = 16, skiprows = [17])
  print("Successfully loaded data")
except FileNotFoundError:
  print("File not found")
Energy.head()
```

```python
Energy = Energy.drop(columns=['Unnamed: 0', 'Unnamed: 1'])
```

```python
Energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
```

```python
Energy = Energy.replace("...", np.nan)
```

```python
Energy["Energy Supply"] /= 1000000
Energy.head()
```

```python
Energy['Country'] = (
    Energy['Country']
    .str.replace(r'\d+', '', regex=True)
    .str.replace(r' \(.*\)', '', regex=True)
    .replace({
        'Republic of Korea': 'South Korea',
        'United States of America': 'United States',
        'United Kingdom of Great Britain and Northern Ireland': 'United Kingdom',
        'China, Hong Kong Special Administrative Region': 'Hong Kong'
    })
)
```

```python
Energy['Country'] = Energy['Country'].replace({
    'Bolivia (Plurinational State of)': 'Bolivia',
    'Switzerland17': 'Switzerland',
})
```

```python
try:
  GDP = pd.read_excel('/content/world_bank.xls', header = 3)
  print("Successfully loaded data")
except FileNotFoundError:
  print("File not found")
GDP.head()
```

```python
GDP['Country Name'] = GDP['Country Name'].replace({
    'Korea, Rep.': 'South Korea',
    'Iran, Islamic Rep.': 'Iran',
    'Hong Kong SAR, China': 'Hong Kong'})
```

```python
ScimEn = pd.read_excel('/content/scimagojr country rank 1996-2024.xlsx')
ScimEn.head()
```

```python
ScimEn = ScimEn.head(15)
ScimEn = ScimEn.drop(columns = ['Region'])
```

```python
GDP = GDP.loc[:, ['Country Name'] + list(map(str, range(2006, 2016)))]
GDP = GDP.rename(columns = {'Country Name': 'Country'})
```

```python
answer_one = pd.merge(ScimEn, Energy, on='Country', how = 'left')
answer_one = pd.merge(answer_one, GDP, on='Country', how = 'left')
answer_one = answer_one.set_index('Country')
```

###Question 2

```python
def answer_two():
    Top15 = answer_one
    years = [str(year) for year in range(2006, 2016)]
    avgGDP = Top15[years].mean(axis=1).sort_values(ascending=False)
    return avgGDP
print(answer_two())
```

###Question 3

```python
def answer_three():
  avgGDP = answer_two()
  country = avgGDP.index[5]
  Top15 = answer_one
  answer = Top15.loc[country]['2015'] - Top15.loc[country]['2006']
  return answer
print(answer_three())
```

###Question 4

```python
def answer_four():
  Top15 = answer_one
  Top15["Self-Citations to Total Citations"] = Top15['Self-citations']/ Top15['Citations']
  country = Top15["Self-Citations to Total Citations"].idxmax()
  max_ratio = Top15.loc[country, "Self-Citations to Total Citations"]
  return (country, max_ratio)
print(answer_four())
```

###Question 5

```python
def answer_five():
    Top15 = answer_one
    Top15["Population"] = Top15["Energy Supply"] * 1000000/ Top15["Energy Supply per Capita"]
    third = Top15["Population"].nlargest(3).index[-1]
    return third
print(answer_five())
```

###Question 6

```python
def answer_six():
    Top15 = answer_one
    Top15['Citable Documents Per Person'] = Top15['Citable documents'] / Top15['Population']
    return Top15['Citable Documents Per Person'].corr(Top15['Energy Supply per Capita'], method='pearson')
print(answer_six())
```

###Question 7

```python
def answer_seven():
    Top15 = answer_one
    ContinentDict  = {
        'China':'Asia',
        'United States':'North America',
        'Japan':'Asia',
        'United Kingdom':'Europe',
        'Russian Federation':'Europe',
        'Canada':'North America',
        'Germany':'Europe',
        'India':'Asia',
        'France':'Europe',
        'South Korea':'Asia',
        'Italy':'Europe',
        'Spain':'Europe',
        'Iran':'Asia',
        'Australia':'Australia',
        'Brazil':'South America'}
    Top15['Continent'] = Top15.index.to_series().map(ContinentDict)
    result = Top15.groupby('Continent')['Population'].agg(
        size='count',
        sum='sum',
        mean='mean',
        std='std')
    return result
print(answer_seven())
```
